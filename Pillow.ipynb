```python
from PIL import Image
from PIL import ImageColor

new=Image.new("RGBA",(500,500),"Yellow")

for x in range(125,375) :
    for y in range(125,250) :
        new.putpixel((x,y),(122,255,215,255)) # only two arqument
new.save("new.png")

for x in range(125,375) :
    for y in range(251,375) :
        new.putpixel((x,y),ImageColor.getcolor("Blue","RGBA"))
        
new.save("new.png") #can not save as .jpg
new
```

```python
from PIL import Image

i=Image.open("test.jpg")
copy1=i.copy()
copy1=copy1.crop((500,500,600,600))
w,h=copy1.size #70,70

new_w,new_h=700,700
new=Image.new("RGB",(new_w,new_h),"Yellow")

for x in range(0,new_w,100):
    for y in range(0,new_h,100):
        new.paste(copy1,(x,y))
copy1
new
```

```python
from PIL import Image
from PIL import ImageFilter

i=Image.open("test.jpg")
#filter_test=i.filter(Image.FILTERED.Blur) AttributeError: 'int' object has no attribute 'Blur'
i=i.filter(ImageFilter.BLUR)
i.save("test_BLUR.jpg")

i=Image.open("test.jpg")
i=i.filter(ImageFilter.CONTOUR)
i.save("test_CONTOUR.jpg")

i=Image.open("test.jpg")
i=i.filter(ImageFilter.EMBOSS)
i.save("test_EMBOSS.jpg")
```

```python
from PIL import Image,ImageDraw

i=Image.new("RGBA",(300,300),"Yellow")
obj=ImageDraw.Draw(i)
for x in range(100,200,3):                    #point
    for y in range(100,200,3):
        obj.point([(x,y)],fill="black")

for x in range(150,300,10):               # 
    obj.line([(x,0),(300,x-150)],fill="Blue")

for y in range(150,300,10) :              #
    obj.line([(0,y),(y-150,300)],fill="Blue")

    
i.save("Draw.png")
```

```python
from PIL import Image,ImageDraw,ImageFont

i=Image.new("RGBA",(300,300),"Yellow")
obj=ImageDraw.Draw(i)

obj.ellipse((50,30,130,70),outline="black")   #eye outline
obj.ellipse((170,30,250,70),outline="black")

obj.ellipse((85,45,95,55),fill="blue")       #eye
obj.ellipse((205,45,215,55),fill="blue")

obj.rectangle((115,240,185,270),fill="red") #mouth

obj.polygon([(150,100),(135,150),(165,150)],fill="cyan")

string="This is Frank"
obj.text((15,90),string,fill="black")

font_calibrii=ImageFont.truetype("C:\Windows\Fonts\calibrii.ttf",24)
obj.text((15,110),string,fill="black",font=font_calibrii)

string_c=""
font_mingliub=ImageFont.truetype("C:\Windows\Fonts\msjhbd.ttc",26)
obj.text((15,130),string_c,fill="black",font=font_mingliub)

i.save("draw_face.png")
```
