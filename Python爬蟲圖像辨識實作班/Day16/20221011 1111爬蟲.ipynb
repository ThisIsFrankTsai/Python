```python
import requests              
from bs4 import BeautifulSoup 
import openpyxl

res = requests.get("https://www.1111.com.tw/search/job?ks=大數據")  #(放URL)
soup = BeautifulSoup(res.text) 
 

wb = openpyxl.Workbook()  #開一個叫wb的空白作業簿
ws = wb.active

ws["A1"] = "職缺名稱"
ws["B1"] = "職缺連結"
ws["C1"] = "公司名稱"
ws["D1"] = "工作地區"
ws["E1"] = "薪資待遇"
ws["F1"] = "給薪方式"
ws["G1"] = "薪資下限"
ws["H1"] = "薪資上限"
```

```python
from time import sleep  #不要爬太快，sleep(1):每頁休息一秒 
page = 1    #while迴圈:符合就會繼續跑
while soup.find_all("div",class_ = "job_item_info")!=[]:  #不符合的時候=讀不到資料 就停
    print("印出第",page,"頁資料")
    print("====================================")
    for job in soup.find_all("div",class_ = "job_item_info"):
        a = (job.h5.text)  #設job抓出這面資料的title，用for迴圈抓出所有資料
        b = (job.a["href"])  #["href"]可以直接抓到連結，["屬性名稱"]=會回傳屬性的內容
        c = (job.h6.text)
        d = (job.find_all("a",class_="job_item_detail_location")[0].text)
        e = (job.find_all("div",class_="job_item_detail_salary")[0].text)

        f = e[:2]  #取前兩個字

        price = ""
        for word in e:
            if word == "0" or word == "1" or word == "2" or word == "3" or word == "4" or word == "5" or word == "6" or word == "7" or word == "8" or word == "9" or word == "萬" or word == "~" or word == ".":
                price += word  #=>price= price +word

        low_price = ""
        high_price = ""
        if "~" in price:
            low_price = price[0:price.find("~")]
            high_price = price[price.find("~")+1:len(price)]  #以上包含，以下不包含，不+1會包含"~"
        else:
            low_price = price
            high_price = price        

        if "萬" in low_price:
            low_price = float(low_price[:low_price.find("萬")])*10000
        else:
            low_price = float(low_price)
        if "萬" in high_price:
            high_price = float(high_price[:high_price.find("萬")])*10000
        else:
            high_price = float(high_price)    

        ws.append([a,b,c,d,e,f,low_price,high_price])
    
    page += 1
    res = requests.get("https://www.1111.com.tw/search/job?ks=大數據&page="+str(page))
    soup = BeautifulSoup(res.text) 
    sleep(1)
    
    wb.save("20221013大數據職缺清單-整理版-1111.xlsx")  #有預設存檔的位置
#0個Tab，存1次；1個Tab，存85次(一頁一次)；2個Tab，存1688次(一職缺一次) => 1個Tab最好，跑多少存多少，萬一當機前面已儲存
```
