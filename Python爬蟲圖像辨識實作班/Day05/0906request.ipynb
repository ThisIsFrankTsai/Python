```python
import requests
from bs4 import BeautifulSoup

data=requests.get("https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A")
soup=BeautifulSoup(data.text)
print(soup.prettify())
```

```python
print(soup.find("h5",class_="card-title title_6").text)
print(soup.find_all("h5",class_="card-title title_6")[0].text)
print(soup.select(".card-title")[0].text)
```

```python
for i in soup.select(".job_item_info"):
    print(i.h5.text)
print("==============================================================================")
for i in soup.find_all("h5",class_="card-title title_6"):
    print(i.text)
```

```python
print(soup.head.prettify()) #列出<head></head>之間的內容
print(soup.head.title.prettify())#列出<head><title></title></head>之間的內容
print(soup.find("title")) #只能找出第一個
print(soup.find_all("title"))#所有符合的內容 list
print(soup.find_all("title")[0])  #抓出0
print(soup.select("title")) #所有符合的內容
print(soup.find_all("title")[0].string) # 抓出最後文字
for i in soup.select("title"): 
    print(i.string)
```
