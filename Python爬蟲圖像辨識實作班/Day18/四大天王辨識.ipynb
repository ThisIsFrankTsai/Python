```python
import cv2
import numpy as np

img=[]
labels=[]
kings=['L','Q','M','Z']

label_num=0

for king in kings:
    num=1
    while num<11:
        img.append(cv2.imread(king+str(num)+'.jpg',cv2.IMREAD_GRAYSCALE))
        labels.append(label_num)
        num+=1
    label_num+=1
print(len(img))
print(len(labels))
print(img)
print(labels)
```

```python
rec=cv2.face.LBPHFaceRecognizer_create()
rec.train(img,np.array(labels))
```

```python
imgtest0=cv2.imread('4kings.jpg')
gray=cv2.cvtColor(imgtest0,cv2.COLOR_BGR2GRAY)

faceCascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces=faceCascade.detectMultiScale( #辨識出人臉所在的座標位置,參數設定至少四行（一張臉一組座標）
    gray,
    scaleFactor=1.2,                #前後兩次搜索框框比例係素（each time小匡掃完換大筐的放大比例->越接近1越細）
    minNeighbors=3,
    minSize=(28,28)
    )

cutlist=[]
num=0
for(x,y,w,h) in faces:
    num+=1
    cuut=imgtest0[y:y+h,x:x+h]
    cv2.imwrite('cuut'+str(num)+'.jpg',cuut)
    cutlist.append('cuut'+str(num)+'.jpg')
print(cutlist)
# print(faces)
```

```python
namelist=['Andy Lau','Aaron Kwok','Lenon Lai','Jacky Cheung']
who=[]
for face in cutlist:
    
    predict_img=cv2.imread(face,cv2.IMREAD_GRAYSCALE)#抽 問你是屬於哪類
    label,confidence=rec.predict(predict_img)
    print('name=',namelist[label])
    print('confidence=',confidence)
    
    who.append(namelist[label])
print(who)
```

```python
num=0
for(x,y,w,h) in faces:               #繪製人臉方匡 參數＝臉的左上點x,y座標,長度,寬度
    cv2.rectangle(imgtest0,(x,y),(x+w,y+h),(0,255,0),3)
    cv2.putText(imgtest0,who[num],(x,y),cv2.FONT_HERSHEY_TRIPLEX,0.8,(0,255,0),2)
    num+=1
cv2.imshow('findface',imgtest0)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
for face in cutlist:
    predict_img=cv2.imread('cuut'+str(num)+'.jpg',cv2.IMREAD_GRAYSCALE)
    label,confidence=rec.predict(predict_img)
    print('label=',label)
    print('confidence=',confidence)
```
