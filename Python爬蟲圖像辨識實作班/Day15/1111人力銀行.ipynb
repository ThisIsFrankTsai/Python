```python
import requests
from bs4 import BeautifulSoup

url="https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A&page=1"

res=requests.get(url)
soup=BeautifulSoup(res.text)

for i in soup.find_all('div','job_item_info'):
    print(i.h5.text)                                            #job title
    print(i.a['href'])                                          #link
    print(i.h6.text)                                            #公司name
    print(i.find('a',class_='job_item_detail_location').text)   #地址
    print(i.find('div',class_='job_item_detail_salary').text)   #salary 
```
