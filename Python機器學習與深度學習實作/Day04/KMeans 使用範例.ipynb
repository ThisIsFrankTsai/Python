```python
#Unsupervised Learning

import numpy as np
import matplotlib.pyplot as plt
```

```python
#產生隨機的資料點

x = np.random.randn(100, 2)
plt.scatter(x[:,0],x[:,1])
```

```python
#使用 KMeans 分群
# 1. 載入想要用的模型
from sklearn.cluster import KMeans
#2. 建立模型
clf = KMeans(n_clusters=3) #分成3群
#3. 訓練模型
clf.fit(x)
clf.labels_

```

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

```python
y = np.random.randn(10, 2)
```

```python
y
```

```python
#4. 使用模型來預測
clf.predict(y)
```

```python
y_target=clf.predict(y)
```

```python
plt.scatter(y[:,0], y[:,1], c=y_target)
```
