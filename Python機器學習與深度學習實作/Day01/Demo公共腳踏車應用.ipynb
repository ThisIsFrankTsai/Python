```python
#新竹市公共自行車租賃系統(YouBike): https://data.gov.tw/dataset/67781

import requests
import csv

# 可以使用 `pip install folium` 安裝套件
# 使用 `pip install geocoder`
import folium 
import geocoder #GPS 座標正查反查

url = "https://odws.hccg.gov.tw/001/Upload/25/OpenData/9059/59/8544b041-b6ab-4b49-bab5-e10237d5c88f.csv"
r = requests.get(url)
#r

# #使用colab讀csv檔前，需要先使用這行來上傳檔案
# 在google colab 上使用才需要
# from google.colab import files
# uploaded = files.upload()

# fn = 'bike02.csv'
# with open(fn, 'r', encoding='big5') as csvfile:
#     rows = csv.reader(csvfile)
#     data_list = list(rows)

#解析 下載下來的csv檔案
decoded_content = r.content.decode('utf-8')
#print(decoded_content)
cr = csv.reader(decoded_content.splitlines(), delimiter=',')
data_list = list(cr)
#data_list[0]
#data_list[1][4] 車站地址
#data_list[1:3]

# # 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新竹市').latlng #open street map 解析地圖功能 lat經度縮寫
location #[24.8066333, 120.9686833] 緯度經度 新竹市中心點
m = folium.Map(location=location, zoom_start=15) #產生map

for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        folium.Marker([float(item[2]), float(item[1])], popup=item[0]+'_'+item[4],
                      icon=folium.Icon(color='blue', 
                                       prefix='fa', icon='fa-arrow-down')).add_to(m)
        #icon 更改marker
        #icon 可以上   https://fontawesome.com/v4/icons/   下載
        #<i class="fa fa-heart" aria-hidden="true"></i>
    except Exception as e:
        print(e.args)    
    
m.save('demo01.html')
m
```
