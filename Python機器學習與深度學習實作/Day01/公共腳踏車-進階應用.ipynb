```python
#新北市公共自行車租賃系統(YouBike): https://data.gov.tw/dataset/123026

import requests
import csv

# 可以使用 `pip install folium` 安裝套件
# 使用 `pip install geocoder`
import folium 
import geocoder #GPS 座標正查反查

url = "https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/csv/file"
r = requests.get(url)
#r

# #使用colab讀csv檔前，需要先使用這行來上傳檔案
# 在google colab 上使用才需要
# from google.colab import files
# uploaded = files.upload()

# fn = 'bike02.csv'
# with open(fn, 'r', encoding='big5') as csvfile:
#     rows = csv.reader(csvfile)
#     data_list = list(rows)

#解析 下載下來的csv檔案
decoded_content = r.content.decode('utf-8')
#print(decoded_content)
cr = csv.reader(decoded_content.splitlines(), delimiter=',')
data_list = list(cr)
#data_list[0]
data_list[1][2] #tot(場站總停車格)
data_list[1][3] #sbi(可借車位數)

# #data_list[1:3]

# # 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新北市').latlng #open street map 解析地圖功能 lat經度縮寫
location #[24.8066333, 120.9686833] 緯度經度 新竹市中心點
m = folium.Map(location=location, zoom_start=15) #產生map

for item in data_list[1:]:
    try:
        # item[6]=緯度, item[7]=經度, item[1]=名稱
        #print(float(item[7]),float(item[6]),item[1],item[3],item[2])
        info='%s 剩餘數量[%s/%s]'%(item[1],item[3],item[2])
        #popup=item[1]+':'+item[3]+'/'+item[2] #點下去才會有資訊
        folium.Marker([float(item[6]), float(item[7])], tooltip=info, #tooltip 不用點就有效果
                      icon=folium.Icon(color='blue', 
                                       prefix='fa', icon='bluetooth')).add_to(m)

    except Exception as e:
        print(e.args)    
    
# m.save('demo01.html')
m
```
