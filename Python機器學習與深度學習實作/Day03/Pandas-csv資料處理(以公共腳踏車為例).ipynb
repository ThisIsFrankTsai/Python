```python
import pandas as pd

url = "https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/csv/file"
df = pd.read_csv(url) 
df.head()
```

```python
# 在colab用底下的安裝指令
import folium
import geocoder

# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新北市').latlng

m = folium.Map(location=location, zoom_start=15)

for i in df.index:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        name = df['sna'][i]  #站名
        total = df['tot'][i] #可借總數量
        sbi = df['sbi'][i]  #剩餘數量
        lat = df['lat'][i]  #緯度
        lng = df['lng'][i]  #經度
        info = '%s 剩餘數量[%s/%s]' %(name, sbi, total)
        folium.Marker([ lat, lng],
                      tooltip=info,
                      popup=df['ar'][i],
                      icon=folium.Icon(color='blue', prefix='fa', icon='fa-bicycle')
                     ).add_to(m)
        
    except Exception as e:
        print(e.args)   
```
